```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.datasets import load_iris

# load datasets
X, y = load_iris(return_X_y=True)
```

```python
from numpy.random import permutation

p = permutation(len(X))

X_r = X[p]
y_r = y[p]
```

```python
dt = DecisionTreeClassifier(max_depth=3)
dt.fit(X_r,y_r)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
%matplotlib inline
fig, ax = plt.subplots(figsize=(20, 20))
plot_tree(dt, filled=True)
```

```python
predictions = dt.predict(X_r)
```

```python
p_t = 0
p_f = 0
for i in range(150):
    if predictions[i]==y_r[i]:
        p_t+=1
    else:
        p_f+=1
        
print(p_t/150)
print(p_f/150)
```

```python
sdt = DecisionTreeClassifier(max_depth=3)
sdt.fit(X_r[:50],y_r[:50])
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
plot_tree(sdt, filled=True)
```

```python
sdt.fit(X_r[50:100],y_r[50:100],update_tree=True)
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
plot_tree(sdt, filled=True)
```

```python
sdt.fit(X_r[100:],y_r[100:],update_tree=True)
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
plot_tree(sdt, filled=True)
```

```python
predictions = sdt.predict(X_r)
```

```python
p_t = 0
p_f = 0
for i in range(150):
    if predictions[i]==y_r[i]:
        p_t+=1
    else:
        p_f+=1
        
print(p_t/150)
print(p_f/150)
```
